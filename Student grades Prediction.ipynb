# Modelling

```python
# IMPORTING MACHINE LEARNING LIBRARIES
from sklearn.neighbors import KNeighborsRegressor  
from sklearn.tree import DecisionTreeRegressor 
from sklearn.linear_model import LogisticRegression  
from sklearn.ensemble import RandomForestRegressor 
from sklearn.tree import ExtraTreeRegressor
from sklearn.svm import SVR
from sklearn.linear_model import Lasso,Ridge
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.33,random_state=49)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

```python
KNR=KNeighborsRegressor()
DTR=DecisionTreeRegressor()
RFR=RandomForestRegressor()
LS=Lasso()
RG=Ridge()

model=[KNR,DTR,RFR,LS,RG]

for m in model:
    m.fit(x_train,y_train)
    predm=m.predict(x_test)
    print("R_2_Score -->",m,r2_score(y_test,predm))
    print("Mean Squared Error -->",m,mean_squared_error(y_test,predm))
    print("Mean Absolute Error -->",m,mean_absolute_error(y_test,predm))
    print("Root Mean Squared Error -->",m,np.sqrt(mean_squared_error(y_test,predm)))
    print('\n')
```

```python
# By Checking R2_Score Ridge and KNeighbour Regressor is our best model
```

```python
# Lets reconfirm our model which is best by using Cross_Validation_Score

from sklearn .model_selection import cross_val_score
```

```python
# Cross-validation for KNeighbors Regressor
scores = cross_val_score(KNR, x_train, y_train, cv=5)
print("Cross-validation scores --> KNR", scores)
print("Mean cross-validation score --> KNR", scores.mean())
print("Difference between R2 Score and CVS -->",(r2_score(y_test,predm)-scores.mean())*100)
print('\n')

# Cross-validation for Decision Tree Regressor
scores = cross_val_score(DTR, x_train, y_train, cv=5)
print("Cross-validation scores --> DTR", scores)
print("Mean cross-validation score --> DTR", scores.mean())
print("Difference between R2 Score and CVS -->",(r2_score(y_test,predm)-scores.mean())*100)
print('\n')


# Cross-validation for Random Forest Regressor
scores = cross_val_score(RFR, x_train, y_train, cv=5)
print("Cross-validation scores --> RFR", scores)
print("Mean cross-validation score --> RFR", scores.mean())
print("Difference between R2 Score and CVS -->",(r2_score(y_test,predm)-scores.mean())*100)
print('\n')

# Cross-validation for GaussianNB
scores = cross_val_score(LS, x_train, y_train, cv=5)
print("Cross-validation scores --> Lasso", scores)
print("Mean cross-validation score --> Lasso", scores.mean())
print("Difference between R2 Score and CVS -->",(r2_score(y_test,predm)-scores.mean())*100)
print('\n')

# Cross-validation for SVC
scores = cross_val_score(RG, x_train, y_train, cv=5)
print("Cross-validation scores --> Ridge", scores)
print("Mean cross-validation score --> Ridge", scores.mean())
print("Difference between R2 Score and CVS -->",(r2_score(y_test,predm)-scores.mean())*100)
print('\n')
```

```python
# Ridge model is very best model among all these models
```
