```python
# List of Categorical Columns
categorical_columns=['job', 'marital', 'education', 'default','housing', 'loan', 'contact','month','poutcome','subscribed']
# Set-Up Subplots with 2 columns to Display 4 Plots in a Single Rows
num_plot=len(categorical_columns)
num_col=4
num_rows=(num_plot//num_col)+(num_plot%num_col>0)

fig,axes=plt.subplots(num_rows,num_col,figsize=(22,20))
fig.tight_layout(pad=60)

# Create bar Plots for Each Categorical Columns
for i, columns in enumerate(categorical_columns):
    row=i//num_col
    col=i%num_col
    ax=axes[row,col]
    sns.barplot(x=bank_train[columns].value_counts().index,
                y=bank_train[columns].value_counts(),ax=ax)
    ax.set_title(f'BarPlot for {columns}')
    ax.set_xticklabels(ax.get_xticklabels(),rotation=65)
    ax.set_xlabel('')

# Hide Empty Subplots in Last Rows
for i in range(num_plot,num_rows*num_col):
    fig.delaxes(axes.flatten()[i])
plt.show()
```

```python
# LIST FOR NUMERICAL COLUMNS
numerical_columns=['age', 'balance','day', 'duration', 'campaign','pdays', 'previous']

# Set-up subplots with 2 columns to display 4 plots in a Single Rows
num_plot=len(numerical_columns)
num_col=4
num_row=int((num_plot + num_col - 1) // num_col)

fig,axes=plt.subplots(num_row,num_col,figsize=(15,5 * num_row))
fig.tight_layout(pad=3.0)

# Create Histplot for Each Numerical Columns
for i, columns in enumerate(numerical_columns):
    row=i//num_col
    col=i%num_col
    ax=axes[row,col]
    sns.histplot(bank_train[columns],kde=True,bins=30,ax=ax,color='blue')
    ax.set_title(f'HistPlot for {columns}')
    ax.set_xlabel(columns)
    ax.set_ylabel('Frequency')
    
# Hide Empty Subplots in Last Rows
for i in range(num_plot,num_row*num_col):
    fig.delaxes(axes.flatten()[i])
plt.show()
```

```python
# Set up subplots
num_plots = len(numerical_columns)
fig, axes = plt.subplots(1, num_plots, figsize=(15, 5))
fig.tight_layout(pad=4.0)

# Create box plots for each numerical column
for i, column in enumerate(numerical_columns):
    ax = axes[i]

    ax.boxplot(bank_train[column])
    ax.set_title(f'Box Plot for {column}')
    ax.set_xlabel(column)

plt.show()
```

```python
outliers_numerical_columns = ['age', 'duration', 'balance', 'campaign']

# Define a function to remove outliers using the IQR method
def remove_outliers(bank_train, column):
    Q1 = bank_train[column].quantile(0.25)
    Q3 = bank_train[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df = bank_train[(bank_train[column] >= lower_bound) & (bank_train[column] <= upper_bound)]
    return df

# Remove outliers for each specified column
for column in outliers_numerical_columns:
    data = remove_outliers(bank_train, column)
```

```python
# Set up subplots
num_plots = len(numerical_columns)
fig, axes = plt.subplots(1, num_plots, figsize=(15, 5))
fig.tight_layout(pad=4.0)

# Create box plots for each numerical column
for i, column in enumerate(numerical_columns):
    ax = axes[i]

    ax.boxplot(data[column])
    ax.set_title(f'Box Plot for {column}')
    ax.set_xlabel(column)

plt.show()
```

In this part of the code Feature Engineering was done which include selection of important Features and Sepration of Target Variables.
and Features and then scaling of Features MAtrix.
THe scaled feature along with the Target Variables then Train and Test Split wuith the Ration 80:20

```python
# List of Features(Excluding the Target Variable)
feature=['age', 'job', 'marital', 'education', 'default', 'balance',
       'housing', 'loan', 'contact', 'day', 'month', 'duration', 'campaign',
       'pdays', 'previous', 'poutcome']

# Seprate the Feature and Target Variables
x=data[feature]
y=data['subscribed']
```

# BASELINE MODEL TRAINING
In this part, The Final training data underwant Machine Learing algorithms which were Random Forest, SVC, Logistic Regression and Decision Tree Classifier and for each of the algorithms their Accuracy, F1-Score, Precision, Recall and Confusion Matrix were output as their results.

# Importing Machine Learning Libraries

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score, confusion_matrix
from sklearn.model_selection import cross_val_score, RandomizedSearchCV
```

```python
# List of Classifier to Evaluate
classifier={
    "Random_Forest":RandomForestClassifier(random_state=42),
    "SCV":SVC(),
    "LogisticRegression":LogisticRegression(random_state=42),
    "DecisionTree":DecisionTreeClassifier(random_state=42)
}

result={}         # Store the Cross_Validation Result

for name,clf in classifier.items():
    clf.fit(x_train,y_train)
    y_pred=clf.predict(x_test)

    accuracy=accuracy_score(y_test,y_pred)
    f1=f1_score(y_test,y_pred)
    Precision=precision_score(y_test,y_pred)
    recall=recall_score(y_test,y_pred)
    cm=confusion_matrix(y_test,y_pred)

    result[name]={
        "Accuracy":accuracy,
        "F1-score":f1,
        "Precision":Precision,
        "Recall_Score":recall,
        "Confusion_matrix":cm
    }
```

```python
# Print Result
for name,metrics in result.items():
    print(f"classifier:{name}")
    print(f"Accuracy:{metrics['Accuracy']:.2f}")
    print(f"F1-score:{metrics['F1-score']:.2f}")
    print(f"Precision:{metrics['Precision']:.2f}")
    print(f"Recall_Score:{metrics['Recall_Score']:.2f}")
    print(f"Confusion_matrix")
    print(metrics['Confusion_matrix'])
    print('\n')
```

# CROSS_VAL_SCORE

```python
for name,clf in classifier.items():
    mean_accuracy=cross_val_score(clf,x_scaled,y,cv=5,scoring='accuracy').mean()
    result[name]=mean_accuracy

# Plot The Accuracy score

plt.figure(figsize=(12,6))
plt.bar(result.keys(),result.values())
plt.xlabel('classifier')
plt.ylabel('mean_accuracy')
plt.title('Classifier Comparision of Cross Validation')
plt.xticks(rotation=45)
plt.show()
```

# HYPER PARAMETER TUNING

```python
rf_param_grid={
    'n_estimators':[100,200,300],
    'max_depth':[None,10,20,30],
    'min_samples_split':[2,5,10,12,16],
    'min_samples_leaf':[1,2,4,8,16],
    'max_features':['auto','sqrt','log2']
}
```

```python
rf_random_search=RandomizedSearchCV(
    RandomForestClassifier
    (random_state=42),
    rf_param_grid,
    n_iter=10,
    scoring='accuracy',
    cv=5,
    random_state=42,
    n_jobs=-1)
```

```python
# Fit the Model with Hyperparameter Tuning
```

```python
rf_random_search.fit(x_train,y_train)
```

```python
best_rf_params = rf_random_search.best_params_
best_rf_accuracy = rf_random_search.best_score_
```

```python
print('Best Random Forest Hyper parameter -->',best_rf_params)
print('\n')
print('Best Random Forest Hyper Parameter accuracy -->',best_rf_accuracy)
```

```python
# Train Random Forest with the Best Hyper Parameter 
```

```python
best_rf=RandomForestClassifier(
    n_estimators= 300, 
    min_samples_split= 2, 
    min_samples_leaf= 1, 
    max_features= 'log2',
    max_depth= 30,
    random_state=42
)
best_rf.fit(x_train,y_train)
y_pred_rf=best_rf.predict(x_test)
```

```python
model_rf={
    'Random Forest':(best_rf,y_pred_rf)
}
```

```python
# Train Random Forest with the best hyperparameters
best_rf = RandomForestClassifier(
    n_estimators=best_rf_params['n_estimators'],
    min_samples_split=best_rf_params['min_samples_split'],
    min_samples_leaf=best_rf_params['min_samples_leaf'],
    max_features=best_rf_params['max_features'],
    max_depth=best_rf_params['max_depth'],
    random_state=42
)
best_rf.fit(x_train, y_train)
y_pred_rf = best_rf.predict(x_test)

# Evaluate the models
models = {
    'Random Forest': (best_rf, y_pred_rf)
}

results = {}
for model_name, (model, y_pred) in models.items():
    accuracy = accuracy_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    cm = confusion_matrix(y_test, y_pred)

    results[model_name] = {
        "Accuracy": accuracy,
        "F1 Score": f1,
        "Precision": precision,
        "Recall": recall,
        "Confusion Matrix": cm,
    }

# Print the evaluation results
for model_name, metrics in results.items():
    print(f"Model: {model_name}")
    print(f"Accuracy: {metrics['Accuracy']:.4f}")
    print(f"F1 Score: {metrics['F1 Score']:.4f}")
    print(f"Precision: {metrics['Precision']:.4f}")
    print(f"Recall: {metrics['Recall']:.4f}")
    print(f"Confusion Matrix:")
    print(metrics['Confusion Matrix'])
    print("\n")
```
